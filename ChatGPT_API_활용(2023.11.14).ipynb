```python
import openai
from openai import OpenAI
import time

# ChatGPT 상호 작용 함수
def chat_with_gpt(prompt):
    client = OpenAI(api_key="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")   # 발급받은 API Key 입력
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    message = response.choices[0].message.content.strip()
    return message

# ChatGPT에 인사하기
print(chat_with_gpt("안녕하세요!"))

# ChatGPT와 대화하기
while True:
    user_input = input("당신: ")
    response = chat_with_gpt("나: " + user_input)
    print("ChatGPT: " + response)
    time.sleep(1)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Self-esteem"
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
content = soup.find(id='mw-content-text').find(class_='mw-parser-output')

# 제목 가져오기
title = soup.find(id="firstHeading").text

# 본문 가져오기
paragraphs = content.find_all('p')
text = ''
for p in paragraphs:
    text += p.text

print(title)
print(text)
```
